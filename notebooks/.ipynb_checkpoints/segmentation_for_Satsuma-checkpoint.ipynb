```python
reference_folder = 'F://Experiments//EMX-triples//20200706//MBOT4_212//Reference_212_CAG_Chrome2s_26dpi_PMTs700_duraFlat//'
# tif_folder ='F://Experiments//EMX-triples//20200217//MBOT2_414//visStim//'
# reference_folder = 'F://Experiments//EMX-triples//20200217//MBOT2_414//Reference_414_CAG_chrimson_11dpi_PMTs650_duraFlat//'
tif = glob(reference_folder + '*.tif')[0]
tif
```

```python
with ScanImageTiffReader(tif) as reader:
        metadata = reader.metadata()
        data = reader.data()

try:
    num_channels = len(metadata.split('channelSave = [')[1].split(']')[0].split(';'))
    fs = float(metadata.split('scanVolumeRate = ')[1].split('\n')[0])
    laser_power = float(metadata.split('nSI.hBeams.powers = ')[0].split('\n')[24][-3:])

except IndexError:
    # in case there is only 1 channel
    num_channels = len(metadata.split('channelSave = [')[0].split(']')[0].split(';'))
    fs = float(metadata.split('scanVolumeRate = ')[1].split('\n')[0])
    laser_power = float(metadata.split('nSI.hBeams.powers = ')[0].split('\n')[24][-3:])

db = {
    'nchannels': num_channels,
    'nplanes': 1,
    'fs': fs,
    'laser_power': laser_power}

print('scanning rate was ' + str(fs))
print('imaging laser power was ' +str(laser_power))
```
