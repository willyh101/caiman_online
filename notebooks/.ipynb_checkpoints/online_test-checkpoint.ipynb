```python
data_path = 'E:/caiman_scratch/ori/out/'
js = glob(data_path + '*.json')[0]
hs = glob(data_path + '*.hdf5')[0]

jdat = load_and_parse_json(js)
cnm = load_as_obj(hs)
```
