<font size="5">COMPUTE THE RELATIVE REDNESS OF THE FOV TO COMPARE ACROSS MICE </font>


steps: 1 get some metadata about the exact laser power and wavelength 2 segment red nuclei with caiman 3 compute the median redness of the ROIs 4 compute the median brightness of the green channel

```python
tif = glob(reference_folder + '*.tif')[-1]
tif
```

```python
#get some metadata about the reference image and the image itself

with ScanImageTiffReader(tif) as reader:
        metadata = reader.metadata()
        data = reader.data()

try:
    num_channels = len(metadata.split('channelSave = [')[1].split(']')[0].split(';'))
    fs = float(metadata.split('scanVolumeRate = ')[1].split('\n')[0])
    laser_power = float(metadata.split('nSI.hBeams.powers = ')[0].split('\n')[24][-3:])

except IndexError:
    # in case there is only 1 channel
    num_channels = len(metadata.split('channelSave = [')[0].split(']')[0].split(';'))
    fs = float(metadata.split('scanVolumeRate = ')[1].split('\n')[0])
    laser_power = float(metadata.split('nSI.hBeams.powers = ')[0].split('\n')[24][-3:])

db = {
    'nchannels': num_channels,
    'nplanes': 1,
    'fs': fs,
    'laser_power': laser_power}

print('scanning rate was ' + str(fs))
print('imaging laser power was ' +str(laser_power))
```

```python
#separate the green and red channels
data_red = np.mean(data[1:-1:2,:,:],axis=0)
data_green = np.mean(data[0:-1:2,:,:],axis=0)
green_median = np.median(data_green)
red_median = np.median(data_red)
#subselect the pixels filtered above
data_red = data_red[:,x_start:x_end]
```

```python
#get the masks for each ROI
dims = (512,x_end-x_start)
masks = np.zeros((Ain.shape[1],512,x_end - x_start))
allMasks = np.zeros((512,x_end - x_start))

for i in range(Ain.shape[1]):
    masks[i] = np.reshape(Ain[:,i], dims, order='F')
    allMasks[masks[i]>0] = 1

#verify that the masks are OK    
plt.imshow(allMasks>0)
```

```python
redness = np.median(data_red[allMasks>0])
relative_red = redness/green_median
```

```python
#compute an histogram of the distributions of intensities for those pixels:

plt.hist(data_red[allMasks>0])
plt.xlabel('intensity')
plt.ylabel('pixel count')
plt.title('intensities distribution')
print('the median redness of the ROIs was ' + str(redness))
print('so, relative to the green channel the redness is '+ str(relative_red))
```
